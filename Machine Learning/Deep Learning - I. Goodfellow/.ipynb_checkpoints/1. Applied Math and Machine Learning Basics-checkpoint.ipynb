```python
import numpy as np
import matplotlib.pyplot as plt
```

# Eigenvectors and Eigenvalues

```python
A = np.array([[0.5,0],[0,2]])
w, v = np.linalg.eig(A)
```

```python
print("The eigenvalues of A:\n", w)
print("The eigenvectors of A:\n", v)
```

The eigenvectors and eigenvalues of a matrix A state the directions and the scale that multiplying a vector x by A will have.

```python
theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)
x_new, y_new = np.dot(A, np.array([x,y]))

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)
ax[1].plot(x_new, y_new)

ax[0].set_aspect('equal')
ax[1].set_aspect('equal')
```
